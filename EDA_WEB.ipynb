#Parte 1 Analisis de vistas (web1)

* 1.Importar los datos (web1)
* 2.Cambiar las columnas
    * dia -> Fecha
    * Visitas de la Págia -> visita_pagina
    * Tiempo promedio en la página (segundos) -> tiempo_seg
* 3.Crear una columna para año y mes
* 4.Verificar si hay nulos
* 5.Crear en un DataFrame nuevo solo 2021 y 2022
* 6.Revisar con un histograma las variables visita_pagina,tiempo_seg y Clicks
* 7.Realizar un ScatterPlot de visita_pagina VS tiempo_seg

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel("/content/drive/MyDrive/1_Clases_Python/8 Octava clase/web_1.xlsx")

df.head(5)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
"""
2.Cambiar las columnas
    * dia -> Fecha
    * Visitas de la Págia -> visita_pagina
    * Tiempo promedio en la página (segundos) -> tiempo_seg:
"""
df = df.rename(columns={'dia':'Fecha'})
df = df.rename(columns={'Visitas de la Página':'visita_pagina'})
df = df.rename(columns={'Tiempo promedio en la página (segundos)':'tiempo_seg'})
df.head(5)
```

```python
"""
3.Crear una columna para año y mes
"""
df['año'] = df['Fecha'].dt.year
df['mes'] = df['Fecha'].dt.month
df.head(5)
```

```python
"""
* 4.Verificar si hay nulos
"""
df.isna().sum()
```

```python
'''
* 5.Crear en un DataFrame nuevo solo 2021 y 2022
'''
df_21_22=df[df['año'].isin([2021, 2022])]
df_21_22.head(5)
```

```python
'''
* 6.Revisar con un histograma las variables visita_pagina,tiempo_seg y Clicks
'''

sns.histplot(
    x=df_21_22['visita_pagina']
)
```

```python
sns.histplot(
    x=df_21_22['tiempo_seg']
)
```

```python
sns.histplot(
    x=df_21_22['Clicks']
)
```

```python
'''
7.Realizar un ScatterPlot de visita_pagina VS tiempo_seg
'''
sns.scatterplot(
    data=df_21_22,
    x='visita_pagina',
    y='tiempo_seg'
)
```

# Parte 2: Analisis de Margenes (web2)

* 1.Importar los datos (web2)
* 2.Calcular el Margen y Rentabilidad
    * Margen -> Ingreso - Gasto
    * Rentabilidad -> Margen/Gasto
    
* 3.Crear una columna para año y mes (similar al anterior).
* 4.Ingresar en un dataFrame solo los años 2021 y 2022 y visualisar agrupado los margenes y rentabilidad por año.
* 4.Hacerlo mismo que 3 pero solo para 2023


```python
df2 = pd.read_excel("/content/drive/MyDrive/1_Clases_Python/8 Octava clase/web_2.xlsx")

df2.head(5)
```

```python
"""
2.Calcular el Margen y Rentabilidad

Margen -> Ingreso - Gasto
Rentabilidad -> Margen/Gasto
"""
df2['Margen']=df2['Ingresos']-df2['Gasto']
df2['Rentabilidad']=df2['Margen']/df2['Gasto']

df2.head(5)
```

```python
'''
* 3.Crear una columna para año y mes (similar al anterior).
'''
df2['año'] = df2['dia'].dt.year
df2['mes'] = df2['dia'].dt.month
df2.head(5)
```

```python
'''
* 4.Ingresar en un dataFrame solo los años 2021 y 2022 y visualisar agrupado los margenes y rentabilidad por año.
'''
df2_filt=df2[df2['año'].isin([2021, 2022])]
df2_filt.head(5)
```

```python
margen_por_año=df2.groupby('año')['Margen'].mean()
print(margen_por_año)
```

```python
ren_por_año=df2.groupby('año')['Rentabilidad'].mean()
print(ren_por_año)
```
